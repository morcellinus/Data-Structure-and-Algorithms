# 1. 배열 원소의 최댓값 구하기

````python
def max_of(a):
    maximum = a[0]
    for i in range(1, len(a)): 
        if a[i] > maximum:
            maximum = a[i]
````

위 for 반복문에서 a[1]부터 a[len(a)]까지 차례대로 주목하며 크기를 비교한다.

이렇게 배열 원소를 하나씩 차례로 주목해서 살펴보는 방식을 '스캔'이라고 한다.

배열의 모든 원소에 대한 스캔을 마치면 배열 a의 최댓값이 maximum에 남아있게 된다.

# 2. 배열 원소의 최댓값을 구하는 함수 구현하기

````python
from typing import Any, Sequence # Any는 제약이 없는 임의의 자료형을 의미하고 Sequence는 시퀀스형(리스트, 바이트, 문자열, 튜플, 바이트열)을 의미한다.

def max_of(a: Sequence) -> Any: # 건너받는 매개변수 a의 자료형은 Sequence이고, 반환하는 것은 임의의 자료형 Any이다.
    '''시퀀스형 a 원소의 최댓값을 반환'''
    maximum = a[0]
    for i in range(1, len(a)):
        if a[i] > maximum:
            maximum = a[i]
    return maximum

if __name__ == '__main__': 
    
    '''파이썬에서는 하나의 스크립트 프로그램을 모듈이라고 한다. 
    확장자를 포함하지 않는 파일의 이름 자체를 모듈 이름으로 사용한다. (ex) 파일 이름이 max.py라면 모듈 이름은 max
    위 if문에서는 __name__과 "__main__"이 같은지를 판단한다. 왼쪽 피연산자 __name__은 모듈 이름을 나타내는 변수이다.
    
    스크립트 프로그램이 직접 실행될 때 변수 __name__은 "__main__"이고,
    스크립트 프로그램이 임포트 될 때 변수 __name__은 원래의 모듈 이름(max)이다.'''
    
    print('배열의 최댓값을 구합니다.')
    num = int(input('원소 수를 입력하세요.: '))
    x = [None] * num
    
    for i in range(num):
        x[i] = int(input(f'x[{i}]값을 입력하세요.: '))
        
    print(f'최댓값은 {max_of(x)}입니다.')
````

# 3. 배열 원소를 역순으로 정렬하기

배열 원소를 역순으로 정렬하는 알고리즘을 생각해보면, 교환 횟수는 원소 개수 // 2번이 된다. (홀수일 경우 가운데 원소는 교환 X)

원소 수가 n인 배열을 고려할 때 i값을 원소의 인덱스라고 한다면,

* 왼쪽 원소의 인덱스 : i
* 오른쪽 원소의 인덱스: n-i-1

따라서 알고리즘을 간단히 나타내면 다음과 같다;

```python
for i in range(n // 2):
    # a[i]와 a[n-i-1]의 값을 교환
```

````python
from typing import Any, MutableSequence

def reverse_array(a: MutableSequence) -> None:
    
    n = len(a)
    for i in range(n//2):
        a[i], a[n-i-1] = a[n-i-1], a[i]
        
if __name__ == '__main__':
    print('배열 원소를 역순으로 정렬합니다.')
    nx = int(input('원소 수를 입력하세요.: '))
    x = [None] * nx
    
    for i in range(nx):
        x[i] = int(input(f'x[{i}]값을 입력하세요.: '))
        
    reverse_array(x)
    
    print('배열 원소를 역순으로 정렬했습니다.')
    
    for i in range(nx):
        print(f'x[{i}] = {x[i]}')
````

# 4. 기수 변환하기 (n진수 구하기)

10진수 정수를 n진수로 변환하기 위해선 정수를 n으로 나눈 나머지를 구하는 동시에 몫을 반복해서 나눠야 한다.

몫이 0이 될 때까지 이 과정을 반복하며 나머지를 역순으로 늘어놓으면 기수로 변환한 수가 된다.

````python
# 10진수 정숫값을 입력받아 2 ~ 36진수로 변환해 출력하기

def card_conv(x: int, r:int) -> str: # 정숫값 x를 r진수로 변환한 뒤, 그 수를 나타내는 문자열을 반환
    
    d = ''
    dchar = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    
    while x > 0:
        d += dchar[x%r] # 해당하는 문자(나머지에 해당하는)를 꺼내 결합
        x //= r # x를 r로 나눈 몫을 x에 다시 저장
        
    return d[: : -1] # 역순으로 나머지(기수 변환 된 수)를 반환

if __name__ == '__main__':
    print('10진수를 n진수로 변환합니다.')
    
    while True:
        while True: # 음이 아닌 정수를 입력받음 (반환 대상이 될 정수)
            no = int(input('변환할 값으로 음이 아닌 정수를 입력하세요.: '))
            if no > 0:
                break
        
        while True: # 2 ~ 36진수의 정숫값을 입력받음 (기수 값)
            cd = int(input('어떤 진수로 변환할까요?: '))
            if 2 <= cd <= 36:
                break
        
        print(f'{cd}진수로는 {card_conv(no, cd)}입니다.')
        
        retry = input('한 번 더 변환할까요? (Y ... 예 / N ... 아니요): ')
        
        if retry in {'N', 'n'}:
            break
````

````python
# 함수 card_conv를 explainable하게 바꿔보자

def card_conv(x: int, r: int) -> str:
    d = ''
    dchar = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    n = len(str(x))
    
    print(f'{r:2} | {x:{n}d}')
    while x > 0: # x의 값이 r로 나눈 몫으로 계속 업데이트 되다가 더 이상 나눠지지 못할 때 while문 종료
        print('   +' + (n+2) * '-')
        if x // r: # 몫이 있을 때
            print(f'{r:2} | {x // r:{n}d} ... {x%r}')
        else:
            print(f'   {x // r:{n}d} ... {x%r}')
        d += dchar[x%r]
        x //= r
    return d[::-1]

if __name__ == '__main__':
    print('10진수를 n진수로 변환합니다.')
    
    while True:
        while True: # 음이 아닌 정수를 입력받음 (반환 대상이 될 정수)
            no = int(input('변환할 값으로 음이 아닌 정수를 입력하세요.: '))
            if no > 0:
                break
        
        while True: # 2 ~ 36진수의 정숫값을 입력받음 (기수 값)
            cd = int(input('어떤 진수로 변환할까요?: '))
            if 2 <= cd <= 36:
                break
        
        print(f'{cd}진수로는 {card_conv(no, cd)}입니다.')
        
        retry = input('한 번 더 변환할까요? (Y ... 예 / N ... 아니요): ')
        
        if retry in {'N', 'n'}:
            break
````

# 5. 소수 나열하기

````python
counter = 0

for n in range(2, 1001):
    for i in range(2, n):
        counter += 1
        if n % i == 0: # 2 이상 자기 자신 미만의 정수에 의해 나누어 떨어지는 경우
            break
    else:
        print(n)
        
print(f'나눗셈을 수행한 횟수: {counter}')
````

위 알고리즘을 사용하면 불필요한 나눗셈도 수행해야 한다.

만약 n이 소수일 경우 for반복문은 끝까지 수행되는데, 이 때 n을 나누는 정수 중 합성수도 있기 때문이다.

실질적으로 소수 판별을 위해 필요한 조건은, n이 2부터 n-1 사이의 어떤 소수로도 나누어 떨어지지 않는다는 조건이다.

이를 반영해서 알고리즘을 개선해보자.

````python
counter = 0
ptr = 0
prime = [None] * 500

prime[ptr] = 2 # 초깃값으로 사용
ptr += 1

for n in range(3, 1001, 2): # 홀수만 대상으로 함 (2보다 큰 짝수는 모두 소수가 아니기 때문)
    for i in range(1, ptr): # 이미 찾은 소수로만 나눗셈을 하자
        counter += 1
        if n % prime[i] == 0:
            break
    else:
        prime[ptr] = n
        ptr += 1
        
for i in range(ptr):
    print(prime[i]) 
    
print(f'나눗셈을 실행한 횟수: {counter}')
````

연산 수행 횟수가 획기적으로 줄었음을 알 수 있다. 그러나 여기서 더 효율적인 알고리즘을 만들 수 있다.

직사각형을 생각해보면 5x20이나 20x5나 동일한 직사각형이다. 즉, 100이 5로 나누어 떨어진다면 20으로도 나누어 떨어진다.

따라서 주어진 정수 n의 제곱근보다 작은 어떠한 소수로도 나누어 떨어지지 않으면 소수라고 판단할 수 있다.

이를 반영해 알고리즘을 개선해보자.

````python
counter = 0
ptr = 0
prime = [None] * 500

prime[ptr] = 2
ptr += 1
prime[ptr] = 3
ptr += 1

for n in range(5, 1001, 2): # 홀수만을 대상으로 설정
    i = 1
    while prime[i] * prime[i] <= n:
        counter += 2
        if n % prime[i] == 0:
            break
        i += 1
    else:
        prime[ptr] = n
        ptr += 1
        counter += 1
        
for i in range(ptr):
    print(prime[i])

print(f'곱셈과 나눗셈을 실행한 횟수: {counter}')
````

````python
# 리스트를 복사하는 copy() 함수는 리스트를 복사한 후 원솟값을 변경하면 복사된 리스트의 원솟값까지 변경될 수 있다.

x = [[1, 2, 3], [4, 5, 6]]
y = x.copy() # x를 y로 얕은 복사
x[0][1] = 9
y
````
